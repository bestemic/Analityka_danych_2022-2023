```python
#b) 
print(df.columns)
house_categorical_df = ['id', 'date', 'waterfront', 'view', 'condition', 'grade', 'zipcode']
house_numerical_df = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'sqft_above',
                      'sqft_basement', 'lat', 'long', 'sqft_living15', 'sqft_lot15', 'yr_built', 'yr_renovated']
```

### Comment
**Numerical variables** <br>
The values of these variables are able to determine the quantities that we can measure. What is more using these kinds of variables, we can observe a relationship between the values.

**Categorical variables** <br>
The values of these variables take values from a specific limited set. What is more, using these types of variables, we can perform a qualitative analysis of data, as well as group data into categories and groups.

Some variables are difficult to unambiguously assign whether they are numeric or categorical. In our dataset, such non-identifiable variables are *bedrooms*, *bathrooms*, *floors*, *grade*, *yr_built* ,*yr_renovated*. For this variables we can categorize the data but also count basic statistics.

```python
#c)
df[house_numerical_df].describe()
```

### Comment
We can notice that *count* isn't meaningful statistic. The most significant and useful statistic turns out to be the *mean*. Looking at the mean for each column, we see that on average a house has 3 bedrooms and 2 bathrooms and costs 540 thousand.

A useful value that describe function returns is also the maximum and minimum of a given column. Thanks to this, we can easily find out that the cheapest house cost 75 thousand and the most expensive 7.7. For our dataset, max also revealed that there may be houses that do not have a natural number of floors.

## Exercise 3
a) Create boxplot of apartment's grade versus it's price. Compare with plot 2b).  
b) Create mosaic plot of any two categorical variables.

```python
#a)
grouped_df = df[['price', 'grade']].groupby('grade')['price'].apply(list)
plt.figure(figsize=(10, 8))
plt.ticklabel_format(axis="y", style='plain')
plt.xlabel("grade")
plt.ylabel("price")
plt.boxplot(grouped_df, flierprops=dict(markeredgecolor='blue', markerfacecolor='blue', marker='o'));
```

### Comment
Observing the graph, we can see that the median increases with the increase in the grade and it is not a linear increase. We also do not have as many outliers for the lowest and highest grades than for grades 5-11. The size between the first and third quantiles also increases as the rating increases.

Comparing to diagram 2b we can see that the shape is similar, in places where in 2b we have crowded information, on the box plot we see "boxes" which inform us that this is where most of the data is. The remaining values are the same as the outliers.

```python
#b)
from statsmodels.graphics.mosaicplot import mosaic

figure, _ = mosaic(df,['bedrooms','floors']);
figure.set_size_inches(10, 8)
figure.get_axes()[0].set_xlabel('bedrooms')
figure.get_axes()[0].set_ylabel('floors')
```

### Comment
Mosaic plots help show relationships and give a visual way to compare groups. We see that houses with one floor and 3 bedrooms are the most popular.
